```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# n = 10^7 m = 10^4
p = np.array([1, 2, 4 ,8, 16])
t = np.array([2.320772, 1.106287, 0.874986, 0.905419, 0.940116])

s = t[0] / t
e = s / p

plt.title("T(P) with N = const")
plt.xlabel("P - number of threads")
plt.ylabel("T, s - process time")
plt.hlines(1.907841, 0, 16, color='g', label='qsort')
plt.plot(p, t, label='parallel merge sort')
plt.legend()
plt.show()

plt.title("S(P) = $T_{1}$/$T_{P}$ with N = const")
plt.xlabel("P - number of threads")
plt.ylabel("S, acceleration")
plt.plot(p, s)
plt.show()

plt.title("E(P) = $S_{P}/P$ with N = const")
plt.xlabel("P - number of threads")
plt.ylabel("E, efficiency")
plt.plot(p, e)
plt.show()
```
